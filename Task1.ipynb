# Wifi Location with Python

```python
import os
import gc
import pandas as pd
import numpy as np # fundamental package for scientific computing with Python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
print(os.listdir("./input"))

wifi_train_df = pd.read_csv('./input/trainingData.csv')
```

```python
print('Size of Wifi train data', wifi_train_df.shape)
```

```python
wifi_train_df.describe()
```

```python
print('Unique values of BUILDINGID', wifi_train_df.BUILDINGID.unique())
print('Unique values of FLOOR', wifi_train_df.FLOOR.unique())
print('Unique values of SPACEID', wifi_train_df.SPACEID.unique())
```

```python
msk = np.random.rand(len(wifi_train_df)) < 0.8
trainDF = wifi_train_df[msk]
testDF = wifi_train_df[~msk]

print('Size of Wifi train data', trainDF.shape)
print('Size of Wifi test data', testDF.shape)
```

```python
cols = [col for col in trainDF.columns if col not in ['FLOOR','BUILDINGID',
                                                      'SPACEID', 'RELATIVEPOSITION',
                                                      'USERID','PHONEID','LATITUDE',
                                                     'LONGITUDE','TIMESTAMP']]
X = trainDF[cols]

y = trainDF['SPACEID']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

```

```python
clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
